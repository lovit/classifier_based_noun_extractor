```python
head = 'l30_r15'
directory = '../data/'
model_fname ='../models/Logistic + L2 (C=1.00) norm l30_r15.pkl'


import pickle    
from py.utils import load_data
from py.noun import TrainedNounExtractor

X, y, x_words, vocabs = load_data(head, directory)
with open(model_fname, 'rb') as f:
    classifier = pickle.load(f)
```

```python
coefficient = {vocabs[j]:coef for j, coef in enumerate(classifier.coef_[0])}
```

```python
from config import sentence_fname
class Sentences:
    def __init__(self, fname):
        self.fname = fname
    def __iter__(self):
        with open(self.fname, encoding='utf-8') as f:
            for doc in f:
                yield doc.strip()

sentences = Sentences('../data/sentences_onlyhangle.txt')
noun_extractor = TrainedNounExtractor(coefficient)
scores = noun_extractor.extract(sentences)
```

```python
lrgraph = noun_extractor.lrgraph
```

```python
sorted_list = sorted(filter(lambda x:not type(x[1]) == tuple and len(x[0]) > 1, scores.items()), key=lambda x:pow(x[1].frequency, 1/2) * x[1].score, reverse=True)[:500]
for noun, score in sorted_list:
    print('%s (s=%.2f, f=%d)' % (noun, score.score, score.frequency))
```
