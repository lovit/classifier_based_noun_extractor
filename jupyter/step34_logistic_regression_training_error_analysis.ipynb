```python
head = 'l30_r15'
directory = '../data/'
model_fname ='../models/Logistic + L2 (C=1.00) norm l30_r15.pkl'


import pickle    
from py.utils import load_data

X, y, x_words, vocabs = load_data(head, directory)
X = X.tocsr()
with open(model_fname, 'rb') as f:
    classifier = pickle.load(f)
```

```python
y_pred = classifier.predict(X)
error_idx = [i for i, pred in enumerate(y_pred) if pred != y[i]]
error_words = [x_words[i] for i in error_idx]
len(error_idx)
```

```python
error_words
```

```python
x_frequency = [v[0] for v in X.sum(axis=1).tolist()]
y_frequency = X.sum(axis=0).tolist()[0]
```

```python
def get_features(l_idx, topk=5):
    features = sorted(X[l_idx,:].nonzero()[1].tolist(), key=lambda x:X[l_idx,x], reverse=True)[:topk]
    features = [(vocabs[i], '%.2f'%classifier.coef_[0,i]) for i in features]
    return features

for idx, word in zip(error_idx, error_words):
    print('%s (%d, answer=%s, prediction=%s)\n   -  %s\n' % (word,
                                          x_frequency[idx],
                                          'N' if y[idx] == 1 else 'V',
                                          'N' if y_pred[idx] == 1 else 'V',
                                          get_features(idx)))
```
