```python
head = 'l30_r15'
directory = '../data/'
model_fname ='../models/Logistic + L2 (C=1.00) norm l30_r15.pkl'


import pickle    
from py.utils import load_data
from py.noun import TrainedNounExtractor

X, y, x_words, vocabs = load_data(head, directory)
with open(model_fname, 'rb') as f:
    classifier = pickle.load(f)
coefficient = {vocabs[j]:coef for j, coef in enumerate(classifier.coef_[0])}
```

```python
from config import chat_sentence_fname
class Sentences:
    def __init__(self, fname):
        self.fname = fname
    def __iter__(self):
        with open(self.fname, encoding='utf-8') as f:
            for doc in f:
                yield doc.strip()

sentences = Sentences(chat_sentence_fname)
noun_extractor = TrainedNounExtractor(coefficient)
scores = noun_extractor.extract(sentences)
```

```python
compounds_ = dict(filter(lambda x:type(x[1]) == tuple, scores.items()))
scores_ = dict(filter(lambda x:type(x[1]) != tuple and x[1].p_eojeol < 0.8, scores.items()))

len(scores), len(scores_), len(compounds_)
```

```python
sorted_list = sorted(filter(lambda x:len(x[0]) > 1, scores_.items()), key=lambda x:pow(x[1].frequency, 1/2) * x[1].score, reverse=True)[:500]
for noun, score in sorted_list:
    print('%s (s=%.2f, f=%d, p_e=%.2f)' % (noun, score.score, score.frequency, score.p_eojeol))
```

```python
def to_csv(scores, fname):
    with open(fname, 'w', encoding='utf-8') as f:
        f.write('\t'.join(['word', 'score', 'frequency', 'p_eojeol']))
        for w, score in sorted(scores.items(), key=lambda x:x[1].frequency, reverse=True):
            if len(w) == 1: continue
            f.write('%s\t%.2f\t%d\t%.4f\n' % (w, score.score, score.frequency, score.p_eojeol))
```
