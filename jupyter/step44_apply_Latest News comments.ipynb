```python
head = 'l30_r15'
directory = '../data/'
model_fname ='../models/Logistic + L2 (C=1.00) norm l30_r15.pkl'


import pickle    
from py.utils import load_data, Sentences
from py.noun import TrainedNounExtractor

X, y, x_words, vocabs = load_data(head, directory)
with open(model_fname, 'rb') as f:
    classifier = pickle.load(f)
coefficient = {vocabs[j]:coef for j, coef in enumerate(classifier.coef_[0])}
```

```python
nouns = {}
for fname in latest_news_comments_fnames:
    noun_extractor = TrainedNounExtractor(coefficient)
    scores = noun_extractor.extract(Sentences(fname))
    nouns[fname.split('/')[-1]] = {'scores': scores,
                                   'frequency': noun_extractor.lset
                                  }
    
    print('%s has %d nouns\n' % (fname.split('/')[-1], len(scores)))
```

```python
topk = 300

for fname, info in nouns.items():
    scores = info['scores']
    scores_ = filter(lambda x:len(x[0]) > 1 and not (48 <= ord(x[0][0]) <= 57),scores.items())
    nouns_in_a_day = list(sorted(filter(lambda x:type(x[1]) != tuple and x[1].p_eojeol < 0.95,scores_), key=lambda x:x[1].frequency, reverse=True))[:topk]
    compound_in_a_day = list(sorted(filter(lambda x:type(x[1]) == tuple,scores_), key=lambda x:min(scores_[x[1][0]].frequency, scores_[x[1][1]].frequency)))
    print('\n\n%s' % fname.split('/')[-1])
    print('n_nouns= %d, n_compounds= %d' % (len(scores), len(compound_in_a_day)))
    print('#nouns:\n%s' % ' '.join([w for w, _ in nouns_in_a_day]))
    
```

```python
from konlpy.tag import Twitter
from collections import Counter
twitter = Twitter()

twitter_nouns = {}
for fname in latest_news_comments_fnames:
    print('\n\n%s' % fname.split('/')[-1])
    with open(fname, encoding='utf-8') as f:
        token_counter = Counter((eojeol for doc in f for eojeol in doc.split()))
    print('n tokens= %d' % len(token_counter))
    noun_counter = {}
    for eojeol, freq in token_counter.items():
        for word, tag in twitter.pos(eojeol):
            if tag == 'Noun':
                noun_counter[word] = noun_counter.get(word, 0) + freq
    print('n nouns= %d' % len(noun_counter))
    twitter_nouns[fname.split('/')[-1]] = noun_counter
```

```python
class Sentences:
    def __init__(self, fnames):
        if type(fnames) == str:
            fnames = [fnames]
        self.fnames = fnames
    def __iter__(self):
        for fname in self.fnames:
            with open(fname, encoding='utf-8') as f:
                for doc in f:
                    yield doc.strip()
                    
noun_extractor_whole = TrainedNounExtractor(coefficient)
scores_whole = noun_extractor_whole.extract(Sentences(latest_news_comments_fnames))
```

```python
topk = 200

for fname, info in nouns.items():
    scores = info['scores']
    scores_ = filter(lambda x:len(x[0]) > 1 and not (48 <= ord(x[0][0]) <= 57),scores.items())
    nouns_in_a_day = dict(filter(lambda x:type(x[1]) != tuple,scores_))
    nouns_with_twitter= twitter_nouns[fname]
    
    print('\n\n%s' % fname.split('/')[-1])
    new_nouns = []
    for i, (noun, score) in enumerate(sorted(nouns_in_a_day.items(), key=lambda x:x[1].frequency, reverse=True)):
        if not (noun in nouns_with_twitter):
            new_nouns.append((noun, score.frequency, i))
        if len(new_nouns) == topk: break

    miss_nouns = []
    for i, (noun, frequency) in enumerate(sorted(nouns_with_twitter.items(), key=lambda x:x[1], reverse=True)):
        if len(noun) == 1: continue
        if not (noun in nouns_in_a_day):
            miss_nouns.append((noun, frequency, i))
        if len(miss_nouns) == topk: break
            
    print('#new nouns\n%s' % '\t'.join(['(%s, %d)' % (w,i) if j%5 != 4 else '(%s, %d)\n' % (w,i) for j, (w,f,i) in enumerate(new_nouns)]).replace('\n\t', '\n'))
    print('#miss nouns\n%s' % ' '.join(['(%s, %d)' % (w,i) if j%5 != 4 else '(%s, %d)\n' % (w,i) for j, (w,f,i) in enumerate(miss_nouns)]).replace('\n\t', '\n'))
```
